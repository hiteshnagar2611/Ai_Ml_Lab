```python
import sklearn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset= pd.read_csv("penguins.csv")
datas=dataset[['Species',	'Island',	'Clutch Completion', 'Culmen Length (mm)',	'Culmen Depth (mm)',	
              'Flipper Length (mm)', 'Body Mass (g)',	'Sex']]
datas
```

```python
spec=datas.Species.value_counts()
spec.plot(kind='pie', autopct="%.2f%%")
```

```python
datas.isnull().sum()
```

```python
datas=datas.dropna()
```

```python
datas.info()
```

```python
datas['Sex'].value_counts()
```

```python
datas.drop(datas.index[(datas['Sex'] == '.')], axis=0, inplace=True) #axis=0 means row
datas['Sex'].value_counts()
```

```python
x=datas[['Species',	'Island',	'Clutch Completion', 'Culmen Length (mm)',	'Culmen Depth (mm)',	
              'Flipper Length (mm)', 'Body Mass (g)']]
#x=data.iloc[:,:-1] all except last
y=datas['Sex']

x2=x.copy()
y2=y.copy()
```

Trying Encoding scheme #1: Label encoding (suitable for ordinal data)

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

sp=x['Species']
sple= le.fit_transform(x['Species'])
sple
```

```python
objs = x.select_dtypes(include = "object").columns
print(objs)
```

```python
for i in objs:
    x[i] = le.fit_transform(x[i])

x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.8, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
#from sklearn.tree import DecisionTreeRegressor    for regression

dte1 = DecisionTreeClassifier(criterion = "entropy")
dte1 = dte1.fit(x_train,y_train)
dte_prediction1 = dte1.predict(x_test)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test, dte_prediction1))
```

```python
from sklearn import tree
plt.figure(figsize=(12,8), dpi=800)
tree.plot_tree(dte1.fit(x_train, y_train)) 
plt.savefig('Decision_tree.png') 
```

```python
dte = DecisionTreeClassifier(criterion = "entropy", max_depth=3)
plt.figure(figsize=(12,8))
tree.plot_tree(dte.fit(x_train, y_train))
plt.savefig('Decision_tree2.png') 
```

```python
from sklearn.tree import plot_tree
fn = list(datas.iloc[:, :-1].columns)
plot_tree(dte, feature_names=fn, filled=True, fontsize=5)
plt.show()
```

Trying Encoding scheme #2: One-Hot encoding (suitable for nominal data)

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()

sp=x2['Species']
spohe= ohe.fit_transform(x2[['Species']])
spohe
```

```python
spo=spohe.toarray()
spo
```

```python
print(ohe.categories_)
```

```python
datas[ohe.categories_[0]] = spo
datas
```

```python
#making task easy with pandas get_dummies()

pd.get_dummies(datas.Species)
```

```python
d1=pd.get_dummies(datas.Species, dtype=int)
d2=pd.get_dummies(datas.Island, dtype=int)
d3=pd.get_dummies(datas['Clutch Completion'], dtype=int)
print(d1)
print(d2)
print(d3)
```

```python
pingu=pd.concat([x2,d1,d2,d3], axis='columns')
pingu
```

```python
pingu=pingu.drop(['Species', 'Island', 'Clutch Completion'], axis=1) #axis=1 means column
pingu
```

```python
x_train, x_test, y_train, y_test = train_test_split(pingu,y2,train_size=0.8, random_state=42)
```

```python
dtg = DecisionTreeClassifier(criterion = "entropy") #deafult
dtg = dtg.fit(x_train,y_train)
dtg_prediction = dtg.predict(x_test)

print(metrics.classification_report(y_test, dtg_prediction))
```
